### Dataset Description
Data is acquired from the ADNI database at adni.loni.usc.edu. 
The ADNI is designed to develop the early detection and tracking of AD.
Baseline 3.0 T MRI scans are selected. 
The scans of each subject are obtained at 6 months, 12 months, 24 months and 36 months.

## Preprocessing 
The T1-weighted images which are provided in the form of 3D Nifti volumes preprocessed into normalized template space. And then segmented into grey matter (GM), white matter (WM), and cerebrospinal fluid (CSF) using the VBM8 toolbox in Suspended Particular Matter (SPM).

```python
def previous_slice():
    pass

def next_slice():
    pass

def process_key(event):
    if event.key == '1':
        previous_slice()
    elif event.key == '2':
        next_slice()




def remove_keymap_conflicts(new_keys_set):
    for prop in plt.rcParams:
        if prop.startswith('keymap.'):
            keys = plt.rcParams[prop]
            remove_list = set(keys) & new_keys_set
            for key in remove_list:
                keys.remove(key)
                
def multi_slice_viewer(volume):
    remove_keymap_conflicts({'1', '2'})
    fig, ax = plt.subplots()
    ax.volume = volume
    ax.index = volume.shape[0] // 2
    ax.imshow(volume[ax.index],cmap = plt.get_cmap(name = 'gray'))
    fig.canvas.mpl_connect('key_press_event', process_key)

def process_key(event):
    fig = event.canvas.figure
    ax = fig.axes[0]
    if event.key == '1':
        previous_slice(ax)
    elif event.key == '2':
        next_slice(ax)
    fig.canvas.draw()

def previous_slice(ax):
    volume = ax.volume
    ax.index = (ax.index - 1) % volume.shape[0]  # wrap around using %
    ax.images[0].set_array(volume[ax.index])

def next_slice(ax):
    volume = ax.volume
    ax.index = (ax.index + 1) % volume.shape[0]
    ax.images[0].set_array(volume[ax.index])




```
